```python
import random
import pygame


# 遊戲初始化
pygame.init()

window_size = [800, 600]
screen = pygame.display.set_mode(window_size)
pygame.display.set_caption("接蘋果遊戲")

background_image = pygame.image.load('./background.png')
screen.blit(background_image, (0, 0))

basket_image = pygame.image.load('./basket.png').convert_alpha()
basket_image_rect = basket_image.get_rect()
basket_width = basket_image_rect.width
basket_height = basket_image_rect.height
basket_x = 400
# 預留間隙 5
basket_y = (600 - basket_height) - 5
screen.blit(basket_image, (basket_x, basket_y))

# 蘋果
apple_image = pygame.image.load('./apple.png').convert_alpha()
apple_rect = apple_image.get_rect()
apple_width = apple_rect.width
apple_height = apple_rect.height
# apple_x = random.randint(0, 800 - apple_width)
apple_x = 400
apple_y = 50
screen.blit(apple_image, (apple_x, apple_y))



# 這個是決定是否跳出遊戲迴圈的變數
is_game_over = False
# 取出起始時間
time_start = pygame.time.get_ticks()
# 文字模組，用來顯示文字，可用來顯示儀表板資料。第一個參數為文字內容（這邊因為初始值為 None），第二個為大小
point_font = pygame.font.SysFont(None, 30)
time_font = pygame.font.SysFont(None, 30)
# 遊戲分數
points = 0
# 總遊戲時間
total_time = 30

# clock = pygame.time.Clock() 

while not is_game_over:
    # clock.tick(30) 
    # 目前時間
    time_now = pygame.time.get_ticks()
    # 取出當下時間，若時間超過 30 秒，則印出 over 30secs 遊戲結束
    during_time = (time_now - time_start) // 1000

    # 蘋果移動
    apple_speed = 3
    apple_y += apple_speed
    screen.blit(apple_image, (apple_x, apple_y))
    
    if basket_x < apple_x < basket_x + basket_width and basket_y < apple_y + apple_height - 50< basket_y+ basket_height:
        points += 5
        apple_x = random.randint(0, 800 - apple_width)
        apple_y = 50
        screen.blit(apple_image, (apple_x, apple_y))
    
    if apple_y >= 600 - apple_height:
        apple_x = random.randint(0, 800 - apple_width)
        apple_y = 50
        screen.blit(apple_image, (apple_x, apple_y))
    
    if during_time >= 30:
        print('over 30 secs')
        is_game_over = True
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            is_game_over = True
        if event.type == pygame.MOUSEBUTTONDOWN:
            mouse_x, mouse_y = pygame.mouse.get_pos()
            # 可以看到取得的是 滑鼠點擊的 x, y 軸 tuple (x, y) 位置
            print(mouse_x, mouse_y)
            # 取得圖片範圍大小的矩形物件，可以想成是圖片物件，可以進一步取得長寬等屬性。搭配初始位置即可判斷是否有點擊到圖片
            
            if mouse_x <= basket_width/2:
                mouse_x = basket_width/2
            if mouse_x >= 800 - basket_width/2:
                mouse_x = 800 - basket_width/2
            
            basket_x = mouse_x - basket_width/2
            screen.blit(basket_image, (basket_x, basket_y))
                   
                

   
    # 繪製圖片
    screen.blit(background_image, (0, 0))
    screen.blit(basket_image, (basket_x, basket_y))
    screen.blit(apple_image, (apple_x, apple_y))
     # 遊戲剩餘時間
    remaining_time = total_time -  during_time
    # 製作遊戲分數文字元素
    point_font_surface = point_font.render('Points: {}'.format(points), True, (0, 0, 0))
    # 繪製遊戲分數文字到視窗
    screen.blit(point_font_surface, (10, 0))
    # 製作剩餘時間文字元素
    time_font_surface = time_font.render('Time: {}'.format(remaining_time), True, (0, 0, 0))
    # 繪製剩餘時間文字到視窗
    screen.blit(time_font_surface, (10, 30))
    
    # 記得更新畫面（有點像是 tkinter 的 update）
    pygame.display.flip()

# 最後記得關閉 pygame
pygame.quit()
```
